```python
import pandas as pd
import os
import glob
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_kmo
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import json
```

```python
def call_all_csv(filename):
       df = pd.read_csv(filename)
       df = df.drop(['frame','timestamp', 'counter'], axis = 1)
       df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
       return df
tempat_dimana_file_berada = "/Users/mdaffarudiyanto/Downloads/data/Preprocessed"
filenames = glob.glob(os.path.join(tempat_dimana_file_berada, "*.csv"))
dataframes = [call_all_csv(df) for df in filenames]

df = pd.concat(dataframes)
df = df.dropna()
df.head()
```

```python
df['factor_1'] = (df['AU01_r'] + df['AU02_r'] + df['AU05_r'])/15
df['factor_2'] = (df['AU04_r'] + df['AU07_r'] + df['AU10_r'] + df['AU25_r'] + df['AU26_r'])/25
df['factor_3'] = (df['AU06_r'] + df['AU12_r'] + df['AU14_r'] + df['AU45_r'])/20
df['factor_4'] = (df['AU09_r'] + df['AU20_r'])/10
df['factor_5'] = (df['AU15_r'] + df['AU17_r'] + df['AU23_r'])/15

df.head()
```

```python
df_filtered = df[['factor_1', 'factor_2', 'factor_3', 'factor_4', 'factor_5', 'label']]
df_filtered.head()
```

```python
index = pd.Index(df_filtered['label'])
index.value_counts()
```

```python
# Normalize only the AU columns (assume they all end with "_r")
au_columns = [col for col in df.columns if col.endswith('_r')]

# Normalize the AU columns by dividing their values by 5
df[au_columns] = df[au_columns] / 5

df.head()
```

```python
df.to_csv("Preprocessed/alldata_17.csv", index=False)
```
