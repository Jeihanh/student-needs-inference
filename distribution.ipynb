```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data for each label
data = {
    "participant_id": [1, 2, 3, 4, 5],
    "I don't need anything right now": [0, 900, 0, 1500, 300],
    "I need an easier question": [600, 300, 0, 300, 0],
    "I need feedback on my progress": [0, 900, 0, 600, 0],
    "I need reminders to stay focused": [281, 0, 300, 300, 139],
    "I need to take a break": [0, 0, 300, 0, 600]
}

# Create the DataFrame
df = pd.DataFrame(data)
df.set_index("participant_id", inplace=True)

# Option 1: Stacked Bar Chart
plt.figure(figsize=(10, 7))
df.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='tab20')
plt.xlabel('Participant ID')
plt.ylabel('Response Value')
plt.title('Distribution of Responses by Participant')
plt.legend(title='Labels', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Option 2: Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df, annot=True, fmt="d", cmap="YlGnBu")
plt.xlabel('Labels')
plt.ylabel('Participant ID')
plt.title('Heatmap of Responses by Participant')
plt.show()

# Option 3: Grouped Bar Chart
df_reset = df.reset_index()
df_melted = df_reset.melt(id_vars='participant_id', var_name='Label', value_name='Value')
plt.figure(figsize=(14, 8))
sns.barplot(data=df_melted, x='participant_id', y='Value', hue='Label')
plt.xlabel('Participant ID')
plt.ylabel('Response Value')
plt.title('Distribution of Responses by Participant and Label')
plt.legend(title='Labels', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame based on your initial data
data = {
    "participant_id": [1, 2, 3, 4, 5],
    "I don't need anything right now": [0, 900, 0, 1500, 300],
    "I need an easier question": [600, 300, 0, 300, 0],
    "I need feedback on my progress": [0, 900, 0, 600, 0],
    "I need reminders to stay focused": [280, 0, 300, 300, 140],
    "I need to take a break": [0, 0, 300, 0, 600]
}

# Create the DataFrame
distribution_df = pd.DataFrame(data)
distribution_df.set_index("participant_id", inplace=True)

# Rename the columns to "Needs 1" to "Needs 5"
new_column_names = {
    "I don't need anything right now": "Needs 1",
    "I need an easier question": "Needs 2",
    "I need feedback on my progress": "Needs 3",
    "I need reminders to stay focused": "Needs 4",
    "I need to take a break": "Needs 5"
}

distribution_df.rename(columns=new_column_names, inplace=True)

# Plot heatmap of class distribution with updated labels
fig, ax = plt.subplots(figsize=(8, 6))
heatmap = sns.heatmap(
    distribution_df,
    annot=True,
    fmt="d",
    cmap="RdPu",  # Matches the "RdPu" colormap
    annot_kws={"size": 12},  # Customize annotation font size
    ax=ax
)

# Customize labels and ticks
heatmap.set_title("Frames distribution of Student Needs labels per participant", fontsize=14)
heatmap.set_xlabel("Classes", fontsize=12)
heatmap.set_ylabel("Participant ID", fontsize=12)
plt.xticks(rotation=45)  # Rotate x-ticks to match the confusion matrix style

plt.show()
```
